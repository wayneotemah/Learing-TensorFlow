# Linear

A straight line function where activation is proportional to input ( which is the weighted sum from neuron ).


```python
# let make a tensor
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python

X = tf.cast(tf.range(-10,10,1),dtype=tf.float32)
plt.plot(X)
```

```python
# plot a linear activation

plt.plot(tf.keras.activations.linear(X))
```

The activation funtion does not change the input values

# ELU activation

Exponential Linear Unit or its widely known name ELU is a function that tend to converge cost to zero faster and produce more accurate results. Different to other activation functions, ELU has a extra alpha constant which should be positive number.


R(z)={z         ,z>0
      α.(ez–1)  ,z<=0}


```python
def elu(x,alpha=1.0):
   return tf.where(x >= 0, x, alpha * (tf.exp(x) - 1))


plt.plot(elu(x=X))
```

# ReLU
A recent invention which stands for Rectified Linear Units. The formula is deceptively simple: max(0,z)
. Despite its name and appearance, it’s not linear and provides the same benefits as Sigmoid (i.e. the ability to learn nonlinear functions), but with better performance.

```python
def relu(x):
  return tf.maximum(0,x)

plt.plot(relu(X))
```

# LeakyReLU
LeakyRelu is a variant of ReLU. Instead of being 0 when z<0
, a leaky ReLU allows a small, non-zero, constant gradient α
 (Normally, α=0.01
). However, the consistency of the benefit across tasks is presently unclear. [1]

```python
def leackyRelu(x,alpha=0.1):
  return tf.maximum(alpha * x,x)

plt.plot(leackyRelu(X))
```

# Sigmoid
Sigmoid takes a real value as input and outputs another value between 0 and 1. It’s easy to work with and has all the nice properties of activation functions: it’s non-linear, continuously differentiable, monotonic, and has a fixed output range.

```python
def sigmoid(x):
  return 1/ (1+ tf.exp(-x))

plt.plot(sigmoid(X))
```

# Tanh
Tanh squashes a real-valued number to the range [-1, 1]. It’s non-linear. But unlike Sigmoid, its output is zero-centered. Therefore, in practice the tanh non-linearity is always preferred to the sigmoid nonlinearity. [1]

```python
def tanh(x):
  return (tf.exp(x)-tf.exp(-x)/tf.exp(x)+tf.exp(-x))

plt.plot(tanh(X))
```
